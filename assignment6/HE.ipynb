```python
# Blood test compatability circuit (depth three)
# (\neg dRh or rRh) and ((\neg dBit0 and \neg dBit1) or (dBit1 and rBit1) or (dBit0 and rBit0))

def bloodtype_compatability_depth3(donor_bt, recipient_bt):
    dRh = donor_bt[2]
    rRh = recipient_bt[2]
    dBit0 = donor_bt[0]
    dBit1 = donor_bt[1]
    rBit0 = recipient_bt[0]
    rBit1 = recipient_bt[1]

    #part1 = (not dRh) or rRh
    #part2 = ((not dBit0) and (not dBit1)) or (dBit1 and rBit1) or (dBit0 and rBit0)
    P = 1 - dRh * (1 - rRh) # Depth 1
    T1 = 1 - (dBit0 + dBit1 - (dBit0 * dBit1)) # Depth 1
    T2 = dBit1 * rBit1 # Depth 1
    T3 = dBit0 * rBit0 # Depth 1
    Q1 = (1 - T1) * (1 - T2) # Depth 2
    Q2 = Q1 * (1 - T3) # Depth 3
    Q = 1 - Q2 # Depth 3
    return Q * P # Depth 4

```
